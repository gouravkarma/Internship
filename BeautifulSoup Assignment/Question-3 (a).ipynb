```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Team=soup.find('span',class_="u-hide-phablet")
```

```python
Team.text
```

```python
Team=[]
for i in soup.find_all('span',class_="u-hide-phablet"):
    Team.append(i.text)    
```

```python
Team
```

```python
a = Team[slice(10)]
a
```

```python
Match=soup.find("td",class_="rankings-block__banner--matches")
Match.text
```

```python
Matches=['23','2714']
for i in soup.find_all("td",class_="table-body__cell u-center-text"):
    Matches.append(i.text)
```

```python
Matches
```

```python
MATCHES=Matches[0:38:2]
MATCHES
```

```python
b = MATCHES[slice(10)]
b
```

```python
POINTS=Matches[1:38:2]
POINTS
```

```python
c = POINTS[slice(10)]
c
```

```python
Rating=soup.find("td",class_="rankings-block__banner--rating")
Rating.text.split('\n')[1]

Rating=['118']
for i in soup.find_all("td",class_="table-body__cell u-text-right rating"):
    Rating.append(i.text)
```

```python
Rating
```

```python
d = Rating[slice(10)]
d
```
