```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options

def get_prime_ministers_data():
    url = "https://www.jagranjosh.com/"

    # Set up Chrome WebDriver
    options = Options()
    options.add_argument("--headless")  # Run Chrome in headless mode (no visible browser window)
    driver = webdriver.Chrome(options=options)

    try:
        # Step 1: Navigate to the webpage
        driver.get(url)
        time.sleep(2)  # Wait for page to load

        # Step 2: Click on the "GK" option
        gk_button = driver.find_element(By.LINK_TEXT, "GK")
        gk_button.click()
        time.sleep(2)

        # Step 3: Click on "List of all Prime Ministers of India"
        pm_link = driver.find_element(By.LINK_TEXT, "List of all Prime Ministers of India")
        pm_link.click()
        time.sleep(2)

        # Step 4: Scrape the data
        data = []
        rows = driver.find_elements(By.XPATH, "//table[@class='table4'][1]/tbody/tr")
        for row in rows:
            columns = row.find_elements(By.TAG_NAME, "td")
            name = columns[0].text
            born_dead = columns[1].text
            term_of_office = columns[2].text
            remarks = columns[3].text
            data.append((name, born_dead, term_of_office, remarks))

    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        driver.quit()

    return data

def create_dataframe(data):
    columns = ["Name", "Born-Dead", "Term of Office", "Remarks"]
    df = pd.DataFrame(data, columns=columns)
    return df

if __name__ == "__main__":
    prime_ministers_data = get_prime_ministers_data()
    df = create_dataframe(prime_ministers_data)
    print(df)
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options

def get_expensive_cars_data():
    url = "https://www.motor1.com/"

    # Set up Chrome WebDriver
    options = Options()
    options.add_argument("--headless")  # Run Chrome in headless mode (no visible browser window)
    driver = webdriver.Chrome(options=options)

    try:
        # Step 1: Navigate to the webpage
        driver.get(url)
        time.sleep(2)  # Wait for page to load

        # Step 2: Type in the search bar '50 most expensive cars'
        search_input = driver.find_element(By.NAME, "q")
        search_input.send_keys("50 most expensive cars")
        search_input.submit()
        time.sleep(2)

        # Step 3: Click on "50 most expensive cars in the world"
        cars_link = driver.find_element(By.PARTIAL_LINK_TEXT, "50 Most Expensive Cars")
        cars_link.click()
        time.sleep(2)

        # Step 4: Scrape the data
        data = []
        rows = driver.find_elements(By.XPATH, "//div[@class='price']/parent::div/preceding-sibling::h3")
        for row in rows:
            car_name = row.text.strip()
            price = row.find_element(By.XPATH, "./following-sibling::div[@class='price']").text.strip()
            data.append((car_name, price))

    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        driver.quit()

    return data

def create_dataframe(data):
    columns = ["Car Name", "Price"]
    df = pd.DataFrame(data, columns=columns)
    return df

if __name__ == "__main__":
    expensive_cars_data = get_expensive_cars_data()
    df = create_dataframe(expensive_cars_data)
    print(df)
```
