```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Start a new instance of Chrome web browser
driver = webdriver.Chrome()

# Open BCCI website
url = "https://www.bcci.tv/"
driver.get(url)

# Find the "International" link and click on it
international_link = driver.find_element(By.LINK_TEXT, "International")
international_link.click()

# Find the "Fixtures" link and click on it
fixtures_link = driver.find_element(By.LINK_TEXT, "Fixtures")
fixtures_link.click()

# Get the list of all fixtures
fixtures = driver.find_elements(By.CLASS_NAME, "fixture-list__item")

for fixture in fixtures:
    match_title = fixture.find_element(By.CLASS_NAME, "fixture__format").text.strip()
    series = fixture.find_element(By.CLASS_NAME, "fixture__description").text.strip()
    place = fixture.find_element(By.CLASS_NAME, "fixture__venue").text.strip()
    date = fixture.find_element(By.CLASS_NAME, "fixture__date").text.strip()
    time = fixture.find_element(By.CLASS_NAME, "fixture__time").text.strip()

    print(f"Match Title: {match_title}")
    print(f"Series: {series}")
    print(f"Place: {place}")
    print(f"Date: {date}")
    print(f"Time: {time}")
    print("-" * 30)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Start a new instance of Chrome web browser
driver = webdriver.Chrome()

# Open GitHub website
url = "https://github.com/"
driver.get(url)

# Find the "Trending" link and click on it
trending_link = driver.find_element(By.LINK_TEXT, "Trending")
trending_link.click()

# Find all the repository containers
repository_containers = driver.find_elements(By.CSS_SELECTOR, '.Box-row')

for container in repository_containers:
    repository_title = container.find_element(By.CSS_SELECTOR, 'h1.h3 a').text.strip()
    
    try:
        repository_description = container.find_element(By.CSS_SELECTOR, 'p.d-inline-block').text.strip()
    except:
        repository_description = "No description provided"
    
    try:
        contributors_count = container.find_element(By.CSS_SELECTOR, 'a.muted-link span').text.strip()
    except:
        contributors_count = "0"
    
    try:
        language_used = container.find_element(By.CSS_SELECTOR, 'span[itemprop="programmingLanguage"]').text.strip()
    except:
        language_used = "N/A"

    print(f"Repository Title: {repository_title}")
    print(f"Repository Description: {repository_description}")
    print(f"Contributors Count: {contributors_count}")
    print(f"Language Used: {language_used}")
    print("-" * 30)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# Start a new instance of Chrome web browser
driver = webdriver.Chrome()

# Open Billboard website
url = "https://www.billboard.com/"
driver.get(url)

# Find the "Charts" link and click on it
charts_link = driver.find_element(By.LINK_TEXT, "Charts")
charts_link.click()

# Find the "Hot 100" link and click on it
hot_100_link = driver.find_element(By.LINK_TEXT, "Hot 100")
hot_100_link.click()

# Scroll to the bottom of the page to load all the songs
body = driver.find_element(By.TAG_NAME, "body")
body.send_keys(Keys.END)

# Wait for a few seconds to load all the songs
driver.implicitly_wait(5)

# Find all the song containers
song_containers = driver.find_elements(By.CLASS_NAME, "chart-element")

for container in song_containers:
    song_name = container.find_element(By.CLASS_NAME, "chart-element__information__song").text.strip()
    artist_name = container.find_element(By.CLASS_NAME, "chart-element__information__artist").text.strip()
    
    try:
        last_week_rank = container.find_element(By.CLASS_NAME, "chart-element__meta.text--center.color--secondary.text--last").text.strip()
    except:
        last_week_rank = "N/A"
    
    try:
        peak_rank = container.find_element(By.CLASS_NAME, "chart-element__meta.text--center.color--secondary.text--peak").text.strip()
    except:
        peak_rank = "N/A"
    
    try:
        weeks_on_board = container.find_element(By.CLASS_NAME, "chart-element__meta.text--center.color--secondary.text--week").text.strip()
    except:
        weeks_on_board = "N/A"

    print(f"Song Name: {song_name}")
    print(f"Artist Name: {artist_name}")
    print(f"Last Week Rank: {last_week_rank}")
    print(f"Peak Rank: {peak_rank}")
    print(f"Weeks on Board: {weeks_on_board}")
    print("-" * 30)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Start a new instance of Chrome web browser
driver = webdriver.Chrome()

# Open the provided URL
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey"
driver.get(url)

# Find the table containing the data
table = driver.find_element(By.CSS_SELECTOR, '.article-body table')

# Find all rows in the table, skipping the header row
rows = table.find_elements(By.TAG_NAME, 'tr')[1:]

for row in rows:
    columns = row.find_elements(By.TAG_NAME, 'td')
    
    book_name = columns[1].text.strip()
    author_name = columns[2].text.strip()
    volumes_sold = columns[3].text.strip()
    publisher = columns[4].text.strip()
    genre = columns[5].text.strip()

    print(f"Book Name: {book_name}")
    print(f"Author Name: {author_name}")
    print(f"Volumes Sold: {volumes_sold}")
    print(f"Publisher: {publisher}")
    print(f"Genre: {genre}")
    print("-" * 30)

# Close the browser
driver.quit()
```
